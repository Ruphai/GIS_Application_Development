### 2.	Decision Support Statements

Copy the following code to your script:

```python
# User input is requested and saved with a variable
userInput = input("Please enter a number")
# the input function always returns a string value 
# therefore a conversion of the input into a number data type is required
# if calculations with the input are planned 


x = float(userInput)
# Decision statements depending on the user input 
if x < 5:
    print ("number less than 5")
elif x > 5:
    print ("number greater than 5")
else:
    print ("number =  5") 
```

#### Important:
- print, if, elif and else have to be written in lowercase. 
- A colon after a decision statement is required and an indentation of the following block of code

#### list loop
loops through the whole list (once)

```python
fcList = ["rivers.shp", "streets.shp", "lakes.shp"]
for eachFC in fcList:
    print (eachFC)
```

#### counted loop

```python
for num in range (3, 7):
    print (num)
# returns the numbers between 3 (included) and 7 (excluded)
```

### 4.	Nested statements

```python
# Definition of a list
mixList = [1999, "January", 2003, "May", 2009, "April", "July", 2011, "January"]
# list loop 
for eachEntry in mixList:
    # Query if the variable is a string (str) or not
    try:
        if (type(eachEntry) == str):
            print (eachEntry)
    except:
        continue
```

```python
# Definition of a list
mixList = [1999, "January", 2003, "May", 2009, "April", "July", 2011, "January"]
# list loop 
for eachEntry in mixList:
    # Query if the variable is a string (str) or not
    #isinstance(test_string, str)
    if (isinstance(eachEntry, str)):
        print (eachEntry)
```

Please note the different indentations: First indentation after the loop statement. After the decision statement an additional indentation is needed, which is only executed if the variable is a string data type.
Important: The variables which are defined within the looping statements are arbitrarily selected (e.g. “eachEntry”). These are the variables, which will hold the changing value within the loop (here: the different single entries of the list). It is recommended to select self-explanatory names.
